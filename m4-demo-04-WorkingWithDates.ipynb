```python
import pandas as pd
import numpy as np
import matplotlib
import datetime

import matplotlib.pyplot as plt
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
aapl = pd.read_csv('datasets/AAPL.csv')

aapl.head()
```

```python
aapl = aapl.sample(frac=1)
aapl.head()
```

```python
aapl.shape
```

```python
aapl = aapl.sample(frac=1)

aapl.reset_index(inplace=True, drop=True)

aapl.head()
```

```python
aapl.info()
```

```python
aapl['Date'] = pd.to_datetime(aapl['Date'])

aapl.head()
```

```python
aapl.info()
```

```python
aapl.shape
```

```python
aapl.sort_values(by='Date', inplace=True)

aapl.head()
```

```python
aapl.reset_index(inplace=True, drop=True)

aapl.head()
```

```python
aapl['Date'] = aapl['Date'].apply(lambda d: d.to_pydatetime())

aapl.info()
```

```python
aapl['Weekday'] = aapl['Date'].apply(lambda d: d.isoweekday())

aapl.head()
```

```python
extract_year = lambda x: datetime.datetime.strptime(str(x), "%Y-%m-%d %H:%M:%S").year
```

```python
aapl['Year'] = aapl['Date'].apply(extract_year)

aapl.sample(10)
```

```python
extract_month = lambda x: datetime.datetime.strptime(str(x), "%Y-%m-%d %H:%M:%S").month
```

```python
aapl['Month'] = aapl['Date'].apply(extract_month)

aapl.sample(10)
```

```python
plt.figure(figsize=(10, 8))

plt.plot(aapl['Date'], 
         aapl['Close'])


plt.xlabel('Date')
plt.ylabel('AAPL stock price')

plt.show()
```

```python
aapl_by_month = aapl.groupby(by='Month').mean()

aapl_by_month
```

```python
plt.figure(figsize=(10, 8))

plt.plot(aapl_by_month.index, 
         aapl_by_month['Close'])


plt.xlabel('Date')
plt.ylabel('AAPL stock price')

plt.show()
```
